### Formulation mathématique du VRP

#### Fonction objective
<p style="text-align: justify;">L'objectif est de minimiser la distance totale parcourue par la flotte de véhicules :
 $min\sum_{k \in K} \sum_{i \in N} \sum_{j \in N} C_{ij} x_{ijk} $</p>

#### Variables de décision
<p style="text-align: justify;">$x_{ijk}$ = 
$\begin{cases} 
1, & \text{si le véhicule se déplace directement du client } i \text{ au client } j \\ 
0, & \text{sinon} 
\end{cases}$ </p>

#### Contraintes

1. **Chaque route se termine au dépôt (n+1)** :
   $ \sum_{i \in N} x_{i, n+1, k} = 1, \quad \forall k \in K $
2. **Chaque route commence depuis le dépôt (0)** :
   $ \sum_{j \in N} x_{0jk} = 1, \quad \forall k \in K $
3. **Conservation des flux de véhicules** :
   $ \sum_{i \in N} x_{ijk} - \sum_{j \in N} x_{ijk} = 0, \quad \forall k \in K $
4. **Capacité des véhicules** :
   $ \sum_{i \in N} d_i \sum_{j \in N} x_{ijk} \leq Q, \quad \forall k \in K $
5. **Non-négativité** :
   $ \ x_{ijk} \geq 0, \quad \forall k \in K $

#### Ensembles et paramètres
- $K$ :  Ensemble des véhicules
- $N$ :  Ensemble des villes / clients
- $Q$ :  Capacité maximale de chaque véhicule
- $d_i$:  Demande des clients / villes $i$
- $C_{ij}$ :  Distance entre le client $i$ et le client $j$
- $[a_i, b_i]$ :  Fenêtres de temps pour le client $i$

## Générateur de matrices d'adjacence

 <p style="text-align: justify;">L'objectif ici est de créer une représentation graphique et numérique des connexions entre les principales villes françaises à l'aide d'une matrice d'adjacence. Chaque ville est associée à un numéro unique et les distances entre les paires de villes sont indiquées par des valeurs numériques dans la matrice. Cette modélisation permet de simuler un réseau de transport hypothétique entre ces villes, facilitant ainsi l'étude des itinéraires possibles et des distances parcourues. En utilisant un générateur de matrices d'adjacence aléatoire, le code simule des connexions potentielles entre les villes avec une densité et une distance maximale spécifiées, offrant ainsi un outil pour explorer différents scénarios de réseau urbain ou de transport. La visualisation graphique du réseau permet de mieux comprendre la structure et les relations entre les villes, aidant ainsi à la planification et à l'optimisation des itinéraires de véhicules ou à d'autres applications analytiques liées aux réseaux urbains.</p>

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def generate_random_adjacency_matrix(n, density, max_distance):
    # Initialisation de la matrice avec des valeurs infinies (pas d'adjacence)
    matrix = np.full((n, n), float('0'))
    
    # Parcours de chaque paire de points
    for i in range(n):
        for j in range(i + 1, n):
            if np.random.rand() < density:
                distance = np.random.randint(1, max_distance + 1)
                matrix[i, j] = distance
                matrix[j, i] = distance  # Matrice symétrique
    np.fill_diagonal(matrix, 0)
    
    return matrix

def draw_graph_from_adjacency_matrix(adjacency_matrix):
    G = nx.Graph()

    n = adjacency_matrix.shape[0]
    for i in range(n):
        for j in range(i + 1, n):
            if adjacency_matrix[i, j] != 0:
                G.add_edge(i, j, weight=adjacency_matrix[i, j])

    pos = nx.spring_layout(G)  # Layout pour la position des nœuds
    edge_labels = nx.get_edge_attributes(G, 'weight')

    plt.figure(figsize=(10, 10))
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=12, font_weight='bold')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    plt.title("Graphique de la matrice aléatoire")
    plt.show()

n = 20  # Nombre de points
density = 0.2  # Densité
max_distance = 100  # Distance maximale entre deux points adjacents

adjacency_matrix = generate_random_adjacency_matrix(n, density, max_distance)
print(adjacency_matrix)

draw_graph_from_adjacency_matrix(adjacency_matrix)
```

## Implementation algo



### Optimisation de la Distribution des Colis : Maximisation de la Valeur Totale et Gestion des Camions
 <p style="text-align: justify;">L'objectif principal de notre algorithme est d'optimiser la distribution des colis entre plusieurs camions en maximisant la valeur totale des colis livrés, tout en respectant la capacité maximale de chaque camion. Pour atteindre cet objectif, l'algorithme utilise une approche où les colis sont générés aléatoirement avec des tailles variant de 0.1 à 1 mètre cube et des valeurs entre 5 et 1000 euros. Chaque colis est évalué selon un rapport valeur/taille, puis les colis sont triés par ordre décroissant de ce rapport pour maximiser la valeur totale livrée par camion. Chaque camion est chargé progressivement jusqu'à sa capacité maximale, en ajoutant les colis qui maximisent la valeur totale sans dépasser la capacité. Les résultats de chaque camion, incluant la valeur totale des colis, l'espace utilisé, le nombre de colis et les détails spécifiques, sont enregistrés dans un dictionnaire pour analyse ultérieure. Enfin, le nombre total de camions nécessaires pour la livraison est calculé et affiché, accompagné d'une vue d'ensemble des résultats par ville, offrant ainsi une mesure claire de l'efficacité de la livraison dans le cadre spécifique du projet.</p>

```python
import pulp
import random
import numpy as np
 
# Définition de la taille du tableau
taille_tableau_livraison = 22
num_colis_total = 1000
 
# Initialisation du tableau avec des zéros
tableau = [0] * taille_tableau_livraison
 
# Générer des nombres aléatoires pour chaque élément
for i in range(taille_tableau_livraison):
    tableau[i] = random.randint(1, 150)
 
# Calcul de la somme actuelle du tableau
somme_actuelle = sum(tableau)
 
# Ajuster un élément aléatoire pour que la somme totale soit 1000
while somme_actuelle != num_colis_total:
    indice_aleatoire = random.randint(0, taille_tableau_livraison - 1)
    difference = num_colis_total - somme_actuelle
    nouvelle_valeur = max(1, min(150, tableau[indice_aleatoire] + difference))
    tableau[indice_aleatoire] = nouvelle_valeur
    somme_actuelle = sum(tableau)
 
# Stocker le tableau final et la somme
colis_a_livrer = {
    "Colis a livrer par ville": tableau,
    "Somme finale": somme_actuelle
}
 
# Paramètres
capacite_camion = 100  # metres cubes (stockage max d'un semi-remorque)
tournee = 1  # compteur de tournee (à incrémenter)
 
# Générer des tailles aléatoires pour les colis entre 0.1 et 1 metre cube
tailles_colis_total = np.random.uniform(0.1, 1, num_colis_total)
 
# Générer des valeurs aléatoires pour les colis entre 5 et 1000 euros
valeurs_colis_total = np.random.uniform(5, 1000, num_colis_total)
 
# Calculer le rapport valeur/taille pour chaque colis
rapport_valeur_taille = valeurs_colis_total / tailles_colis_total
 
# Regrouper les colis avec leur taille, valeur et rapport valeur/taille
colis = list(zip(tailles_colis_total, valeurs_colis_total, rapport_valeur_taille))
 
# Dictionnaire pour stocker les informations de chaque camion
camions = {}
 
while colis:
    # Trier les colis par rapport au ratio valeur/taille décroissant
    colis_trie = sorted(enumerate(colis, 1), key=lambda x: x[1][2], reverse=True)
 
    # Initialiser les variables pour le suivi des colis
    camion_actuel = []
    espace_utilise = 0
    valeur_totale = 0
 
    # Ajouter des colis au camion jusqu'à atteindre sa capacite
    indices_a_supprimer = []
    for idx, (taille, valeur, _) in colis_trie:
        if espace_utilise + taille <= capacite_camion:
            camion_actuel.append((idx, taille, valeur))
            espace_utilise += taille
            valeur_totale += valeur
            indices_a_supprimer.append(idx - 1)
 
    # Supprimer les colis livres de la liste originale
    colis = [colis[i] for i in range(len(colis)) if i not in indices_a_supprimer]
 
    # Stocker les informations du camion dans le dictionnaire
    camions[tournee] = {
        'valeur totale (en euros)': valeur_totale,
        'espace utilise (en metres cubes)': espace_utilise,
        'nombre colis': len(camion_actuel),
        'colis': camion_actuel
    }
    tournee += 1
 
# Détermine le nombre de camions dont nous avons besoin pour les livraisons
resultat_camions = camions
nombre_de_camions = len(resultat_camions)
 
# Affichage des données de livraison
print(f"{colis_a_livrer}\n")
 
print(f"Nombre de camions utilises : {nombre_de_camions}\n")
```

## Optimisation du Problème de Routage de Véhicules (VRP) : Utilisation de la Programmation Linéaire Mixte (PLM) avec Pulp et Gurobi des villes francaises
 <p style="text-align: justify;">Notre code  implémente notre solution au Problème de Routage de Véhicules (VRP) en utilisant la programmation linéaire mixte (PLM) avec Pulp et le solveur Gurobi. Il vise à minimiser le coût total des trajets de livraison entre différentes villes pour plusieurs camions, en partant et en revenant à Paris tout en visitant chaque ville exactement une fois, sauf Paris. Le programme utilise une matrice des coûts pour représenter les distances entre les villes, en remplaçant les trajets impossibles par une grande constante. Chaque itération du programme calcule un trajet optimal pour un camion, en respectant les contraintes de capacité et d'évitement de sous-tours grâce aux variables de décision et aux contraintes MTZ. Les résultats sont affichés en indiquant les villes visitées dans l'ordre pour chaque camion, offrant ainsi une solution efficace pour optimiser les opérations de livraison dans un environnement logistique.</p>

```python
import pulp

def prog(matrice):  
    VILLES = [
        "Paris", "Lille", "Rouen", "Rennes", "Brest", "Nantes", "Bordeaux",
        "Clermont Ferrand", "Dijon", "Lyon", "Grenoble", "Metz", "Nancy",
        "Strasbourg", "Saint Etienne", "Marseille", "Montpellier", "Nice",
        "Toulon", "Orleans", "Tours", "Toulouse"
    ]

    n = len(VILLES)  
           
    # Matrice des coûts  
    C = matrice  
           
    # Supposons que les trajets impossibles soient marqués par des zéros (en dehors de la diagonale)  
    GRANDE_CONSTANTE = 1e6  
    for i in range(n):  
        for j in range(n):  
            if i != j and C[i][j] == 0:  
                C[i][j] = GRANDE_CONSTANTE  # Remplace les zéros par une grande constante  
    
    for iteration in range(nombre_de_camions):  
        # Modèle  
        prob = pulp.LpProblem("VRP", pulp.LpMinimize)  
               
        # Variables de décision  
        x = pulp.LpVariable.dicts("x", (range(n), range(n)), 0, 1, pulp.LpBinary)  
        u = pulp.LpVariable.dicts("u", range(1, n), 0, n-1, pulp.LpInteger)  
               
        # Fonction objective  
        prob += pulp.lpSum(C[i][j] * x[i][j] for i in range(n) for j in range(n))  
               
        # Contraintes  
        prob += pulp.lpSum(x[0][j] for j in range(1, n)) == 1  # Un seul départ du dépôt  
        prob += pulp.lpSum(x[i][0] for i in range(1, n)) == 1  # Un seul retour au dépôt  
           
        for i in range(1, n):  
            prob += pulp.lpSum(x[i][j] for j in range(n) if j != i) >= 1  # Chaque ville est visitée une fois  
            prob += pulp.lpSum(x[j][i] for j in range(n) if j != i) >= 1  # Chaque ville est quittée une fois  
            
        # Contraintes MTZ pour éviter les sous-tours  
        for i in range(1, n):  
            for j in range(1, n):  
                if i != j:  
                    prob += u[i] - u[j] + (n-1) * x[i][j] <= n-2  
    
        # Résolution du problème avec Gurobi  
        prob.solve(pulp.GUROBI(msg=False))  
           
        # Suivi du trajet  
        current_city = 0  # Commence à Paris  
        trajet = ["Paris"]  
        while True:  
            for j in range(n):  
                if pulp.value(x[current_city][j]) == 1:  
                    trajet.append(VILLES[j])  
                    current_city = j  
                    break
            if current_city == 0:  # Retour à Paris  
                break
           
        # Affichage du trajet  
        print(f"Trajet emprunte lors de l'iteration {iteration + 1} :")  
        print(" -> ".join(trajet))
```

Après avoir visualisé les positions des villes sur une carte, nous avons formulé le problème d'optimisation avec pour objectif de minimiser la distance totale parcourue. Des variables de décision indiquent les connexions entre les villes, tandis que des variables auxiliaires sont utilisées pour éviter les sous-tours. Les contraintes garantissent que chaque ville est visitée exactement une fois, sauf Paris qui peut avoir plusieurs connexions en raison du nombre de camions. Le solveur utilise PuLP, CBC (Coin-or branch and cut), qui résout le problème. Les résultats sont extraits pour reconstruire les tournées complètes, affichées et visualisées sur une carte.

```python
from pulp import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

# Liste des villes
villes = [
    'Paris', 'Marseille', 'Lyon', 'Toulouse', 'Nice', 'Nantes', 
    'Strasbourg', 'Montpellier', 'Bordeaux', 'Lille', 'Rennes', 'Reims', 
    'LeHavre', 'Saint-Étienne', 'Toulon', 'Amiens', 'Metz', 'Besançon', 
    'Orléans', 'Mulhouse', 'Rouen', 'Caen', 'Nancy', 'Tours'
]

# Chargement des données depuis les fichiers CSV
latlng = ['latitude', 'longitude']
positions = pd.read_csv('./position.csv', index_col="City")
temps_vol = pd.read_csv('./flight_time.csv', index_col="City")
distances_df = pd.read_csv('./distance.csv', index_col="City")

# Affichage des positions des villes sur une carte
positions_villes = dict((ville, (positions.loc[ville, 'longitude'], positions.loc[ville, 'latitude'])) for ville in villes)
for ville in positions_villes:
    pos = positions_villes[ville]
    plt.plot(pos[0], pos[1], 'o')
    plt.text(pos[0] + .01, pos[1], ville, horizontalalignment='left', verticalalignment='center')
    
plt.gca().axis('off')

# Calcul des distances entre les villes à partir du dataframe des distances
distances = dict(((v1, v2), distances_df.loc[v1, v2]) for v1 in positions_villes for v2 in positions_villes if v1 != v2)

# Nombre de camions
K = 6

# Création du problème d'optimisation
prob = LpProblem("VRP", LpMinimize)

# Variable indicatrice si la ville i est connectée à la ville j dans la tournée
x = LpVariable.dicts('x', distances, 0, 1, LpBinary)
# Variables auxiliaires pour éliminer les sous-tournées
u = LpVariable.dicts('u', villes, 0, len(villes) - 1, LpInteger)

# Définition de l'objectif : minimiser le coût total (distance totale parcourue)
coût = lpSum([x[(i, j)] * distances[(i, j)] for (i, j) in distances])
prob += coût

# Contraintes pour garantir que chaque ville (sauf Paris) est visitée exactement une fois
for ville in villes:
    cap = 1 if ville != 'Paris' else K
    # Connexion entrante : chaque ville doit avoir exactement une connexion entrante (sauf Paris qui peut avoir K connexions)
    prob += lpSum([x[(i, ville)] for i in villes if (i, ville) in x]) == cap
    # Connexion sortante : chaque ville doit avoir exactement une connexion sortante (sauf Paris qui peut avoir K connexions)
    prob += lpSum([x[(ville, i)] for i in villes if (ville, i) in x]) == cap

# Élimination des sous-tournées avec des contraintes supplémentaires
N = len(villes) / K
for i in villes:
    for j in villes:
        if i != j and (i != 'Paris' and j != 'Paris') and (i, j) in x:
            prob += u[i] - u[j] <= (N) * (1 - x[(i, j)]) - 1

# Résolution du problème d'optimisation
prob.solve()
print(LpStatus[prob.status])

# Récupération des arêtes non nulles (c'est-à-dire les connexions utilisées dans la solution)
arêtes_non_nulles = [e for e in x if value(x[e]) != 0]

def obtenir_prochaine_ville(parent):
    '''Fonction auxiliaire pour obtenir l'arête suivante à partir d'une ville donnée'''
    arêtes = [e for e in arêtes_non_nulles if e[0] == parent]
    for e in arêtes:
        arêtes_non_nulles.remove(e)
    return arêtes

# Création des tournées à partir de Paris
tournées = obtenir_prochaine_ville('Paris')
tournées = [[e] for e in tournées]

# Construction des tournées complètes
for t in tournées:
    while t[-1][1] != 'Paris':
        t.append(obtenir_prochaine_ville(t[-1][1])[-1])

# Affichage des tournées
for t in tournées:
    print(' -> '.join([a for a, b in t] + ['Paris']))

# Calcul du temps total de vol pour la tournée la plus longue
temps_total = 0
for t in tournées:
    temps = 0
    for i in range(len(t)):
        temps += temps_vol.loc[t[i][0], t[i][1]]
    if temps > temps_total:
        temps_total = temps
print(temps_total)

# Dessin des tournées sur la carte
couleurs = [np.random.rand(3) for _ in range(len(tournées))]
for t, c in zip(tournées, couleurs):
    for a, b in t:
        p1, p2 = positions_villes[a], positions_villes[b]
        plt.plot([p1[0], p2[0]], [p1[1], p2[1]], color=c)

# Affichage de la carte avec les tournées
for ville in positions_villes:
    pos = positions_villes[ville 
    plt.plot(pos[0], pos[1], 'o')
    plt.text(pos[0] + .01, pos[1], ville, horizontalalignment='left', verticalalignment='center')

plt.title('%d ' % K + 'personnes' if K > 1 else 'personne')
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.show()

# Affichage du temps total de la tournée la plus longue et de la distance totale
print('Temps total le plus long:', temps_total, '(min)')
print('Distance totale:', value(prob.objective), '(km)')
```
